## `jarvis_march_all` Examples

We will not show the `jarvis_march` function as there is not much to show, and all the capabilities of the function can be shown in `jarvis_march_all`.

The first 4 examples are taken from previous assessments, where we can see that the algorithm works, and chooses to ignore co-linear points, as well as output the convex hull as a list of line segments.

We also show how the function works with a list of numeric vectors, a numeric vector and a data frame.

```r
line<-list(c(1,1),c(2,2),c(3,3))
pgram<-list(c(1,1),c(2,1),c(3,2),c(2,2),c(2,1.5),c(2.5,2.5))
square<-list(c(1,1),c(2,1),c(2,2),c(1,2),c(1.5,1.5),c(1.5,2.5))
random<-rnorm(400)
df<-data.frame(rnorm(200),runif(200))
```

```r
jarvis_march_all(line,plot=TRUE)
```

```r
jarvis_march_all(pgram,plot=TRUE)
```

```r
jarvis_march_all(square,plot=TRUE)
```

```r
jarvis_march_all(random,plot=TRUE)
```

```r
jarvis_march_all(df,plot=TRUE)
```

### Error checking functionality

The function checks for the following errors, inputting the correct R class type and the length of points. We shall now test this:

```r
# List where an element is the wrong size
wronglist<-list(c(1,2),c(3,4),c(4,0),c(1))
jarvis_march_all(wronglist)
```

```r
# Vector of vectors where there isnt an even amount of points
wrongvec<-c(c(1,2),c(3,4),c(4,0),c(1))
jarvis_march_all(wrongvec)
```

Data frames must have columns of the same length, so error checking for the length didn't seem to make sense for me.

```r
#String/character input
str<-c("1","2","3","4","5","6")
jarvis_march_all(str)
```

## `in_hull` Examples

We now show the following examples to show how `in_hull` works and how the print option can be changed. We input our `Original_Set` of points in which we wish to find the convex hull and then test to see if all of our `Test_Set` points are in the hull.

```r
#Are all the points in the line example in it's own hull
in_hull(line,line)
```

```r
#How about the convex hull of pgram being in the convex hull of pgram?
pghull<-jarvis_march_all(pgram)
X<-c()
Y<-c()
for(i in 1:length(pghull)){
    X[i] <- pghull[[i]][1] 
    Y[i] <- pghull[[i]][2] 
}
pghull_final<-data.frame(X,Y)
in_hull(pgram,pghull_final)
```

```r
#Is a single point in the square hull?
in_hull(square,c(1,1))
```

```r
#is the print being annoying?
randhull<-jarvis_march_all(df)
Xs<-c()
Ys<-c()
for(i in 1:length(randhull)){
    Xs[i] <- randhull[[i]][1] 
    Ys[i] <- randhull[[i]][2] 
}
randhull_final<-data.frame(Xs,Ys)
in_hull(df,randhull_final)
```

```r
#we can turn the print off
in_hull(df,randhull_final,print=FALSE)
```

There is also an inbuilt error checker on the `Test_Set` input. It does exactly the same error checking as the `jarvis_march_all` function.

```r
in_hull(line,c(0))
```

```r
in_hull(line,c("1","1"))
```
